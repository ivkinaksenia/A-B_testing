```python
def AB_test_avt(r1, r2):
    s, p - ttest_ind(r1, r2)
    print("Среднее первого рода - ",r1.mean())
    print("Среднее второго рода - ",r2.mean())
    print()
    print("Значение стат.критерия - ",s)
    print("Значение p_value - ",p)
    print()
    return s, p
```
